```python
import numpy as np
import corner
import matplotlib.pyplot as plt

sbi_samples = np.load("../data/results/sbi_posterior.npy")
standax_samples = np.load("../data/results/standax_posterior.npy")

names = ["alpha_low", "alpha_high", "beta", "offset", "step"]

alpha_low_ = -0.271
alpha_high_ = -0.083
beta_ = 3.31
mabs_ = -19.3+19.3
gamma_ = 0.175

truths = [
    alpha_low_,
    alpha_high_,
    beta_,
    mabs_,    # offset
    gamma_    # step
]

fig = corner.corner(
    sbi_samples,
    labels=names,
    truths=truths,
    color="#A31621",
    levels=(0.68, 0.95),
    fill_contours=True,
    plot_density=True,
    plot_datapoints=False,
    smooth=1.0
)

corner.corner(
    standax_samples,
    labels=names,
    color="#1F487E",
    levels=(0.68, 0.95),
    fill_contours=False,   # important: avoid overfilling
    plot_density=True,
    plot_datapoints=False,
    smooth=1.0,
    fig=fig               # overlay on same figure
)

plt.show()
```
