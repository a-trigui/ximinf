```python
from getdist import plots, MCSamples
import numpy as np

sbi_samples = np.load("../data/results/sbi_posterior.npy")
standax_samples = np.load("../data/results/standax_posterior.npy")
cosmologix_samples = np.load("../data/results/cosmologix_posterior.npy")

names = ["alpha", "beta", "offset", "step"]

# Correct: chain names go in `label`, parameter names go in `labels`
gds_sbi = MCSamples(
    samples=sbi_samples,
    names=names,
    labels=names,
    label="SBI"
)

gds_standax = MCSamples(
    samples=standax_samples,
    names=names,
    labels=names,
    label="StandAX"
)

gds_cosmo = MCSamples(
    samples=cosmologix_samples,
    names=names,
    labels=names,
    label="CosmologIX"
)

g = plots.getSubplotPlotter()

g.triangle_plot(
    [gds_sbi,  gds_standax],
    filled=True,
    contour_levels=[0.68, 0.95],
    colors=["#A31621", "#1F487E", "#C07835"]
)
```
