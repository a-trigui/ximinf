# EXAMPLE

```python
from getdist import plots, MCSamples
import numpy as np

sbi_samples = np.load("../data/results/sbi_posterior.npy")
standax_samples = np.load("../data/results/standax_posterior.npy")
cosmologix_samples = np.load("../data/results/cosmologix_posterior.npy")

names = ["alpha", "beta", "offset", "step"]

# Wrap each chain in an MCSamples object
gds_sbi    = MCSamples(samples=sbi_samples, names=names, labels=names)
gds_standax = MCSamples(samples=standax_samples, names=names, labels=names)
gds_cosmo = MCSamples(samples=cosmologix_samples, names=names, labels=names)

# Create a single plotter
g = plots.getSubplotPlotter()

# Triangle plot with all three chains
g.triangle_plot(
    [gds_standax, gds_cosmo, gds_sbi],  # list of MCSamples
    filled=True,
    contour_levels=[0.68, 0.95],
    colors=["#A31621", "#1F487E", "#C07835"],  # one color per chain
)
```
