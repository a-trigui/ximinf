# EXAMPLE

```python
import numpy as np
import corner
import matplotlib.pyplot as plt

# Generate synthetic correlated Gaussian data
mean = np.array([0.0, 1.0])
cov = np.array([[1.0, 0.8],
                [0.8, 1.0]])

samples = np.random.multivariate_normal(mean, cov, size=5000)

figure = corner.corner(
    samples,
    labels=[r"$x$", r"$y$"],
    show_titles=True,
    title_fmt=".2f"
)

plt.show()
```
